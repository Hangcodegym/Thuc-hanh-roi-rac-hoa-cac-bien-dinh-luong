### 1. Khai báo thư viện cần dùng

```python
import pandas as pd
import numpy as np
```

### 2. Tạo biến một danh sách tuổi

```python
ages = [20, 22, 25, 27, 21, 23, 37, 31, 61, 45, 41, 32]
```

```python
# Hãy chia chúng thành các nhóm từ 18 đến 25, 26 đến 35, 36 đến 60 và cuối cùng là 61 trở lên. 
# định nghĩa khoảng giá trị các nhóm
bins = [18, 25, 35, 60, 100]
```

```python
# thực hiện rời rạc hóa
cats = pd.cut(ages, bins)
cats
```

```python
# lấy ra index của nhóm tương ứng với các phần tử
cats.codes
```

```python
# lấy ra các nhóm
cats.categories
```

```python
# thống kê số lượng phần tử ở mỗi nhóm
pd.value_counts(cats)
```

```python
# Thay đổi phía bị đóng bằng cách truyền vào tham số right = False:
pd.cut(ages, [18, 26, 36, 61, 100], right=False)
```

```python
# danh sách nhãn
group_names = ['Youth', 'YoungAdult', 'MiddleAged', 'Senior']
pd.cut(ages, bins, labels=group_names)
```

### 3. sinh dữ liệu ngẫu nhiên gồm 20 phần tử

```python
data = np.random.rand(20)
data
```

```python
cut_data = pd.cut(data, 4, precision=2) # precision=2 giới hạn độ chính xác ở 2 chữ số thập phân
cut_data
```

```python
# lấy ra các nhóm
cut_data.categories
```

```python
# thống kê số lượng phần tử ở mỗi nhóm
pd.value_counts(cut_data)
```

### 4. Sinh ngẫu nhiễn 1000 điểm dữ liệu

```python
data = np.random.randn(1000)
data
```

```python
# qcut sử dụng các lượng tử mẫu thay thế, nên chúng sẽ nhận được các nhóm có kích thước gần bằng nhau
# thực hiện hàm qcut trên dữ liệu vừa sinh ra
cats = pd.qcut(data, 4)
cats
```

```python
cats.categories
```

```python
#thống kê số lượng phần tử
pd.value_counts(cats)
```

```python
# chuyển các lượng tử của riêng mình (các số từ 0 đến 1):
pd.qcut(data, [0, 0.1, 0.5, 0.9, 1.])
```
